## 1. Методами морфологічної обробки збільшити та зменшити контури зображення
1.1 Завантажити та відобразити зображення "1_8.png" директорії "./examples/"

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("/content/1_8.png", cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(img); plt.title('1_8.png'); 
plt.show();
```

1.2 Створити бінарне зображення (зображення із двома значеннями яскравості пікселів - мінімальним та максимальним) методом порогової обробки. Порогове значення яскравості встановити 10.

```python
bin_img = img.copy()
threshold = 10
max_val = 255
min_val = 0
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val
```

1.3 Відобразити бінарне зображення

```python
plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(bin_img,cmap='gray'); plt.title('Бінарне зображення'); 
plt.show();
```

1.4 Встановити ядро розміром 2х2

```python
kernel = np.ones((2,2),np.uint8)
```

1.5 Застосувати ерозію до бінарного зображення та відобразити його

```python
erosion = cv2.erode(bin_img,kernel,iterations = 1)

plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(erosion,cmap='gray'); plt.title('Ерозія'); 
plt.show();
```

1.6 Застосувати діляцію до бінарного зображення та відобразити його

```python
dilation = cv2.dilate(bin_img,kernel,iterations = 1)

plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(dilation,cmap='gray'); plt.title('Діляція'); 
plt.show();
```

## 2. Методами морфологічної обробки підготувати зображення до подальшої обробки
2.1 Завантажити та відобразити зображення "street.jpg" директорії "./examples/"

```python
img_st = cv2.imread("/content/street.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(img_st); plt.title('street_jpg'); 
plt.show();
```

2.2 Створити бінарне зображення (зображення із двома значеннями яскравості пікселів - мінімальне та максимальне) методом порогової обробки. Порогове значення яскравості встановити 170.

```python
bin_img_st = img.copy()
threshold = 170
max_val = 255
min_val = 0
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img_st[row][col] = max_val
        else:
            bin_img_st[row][col] = min_val
```

2.3 Відобразити бінарне зображення

```python
plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(bin_img_st,cmap='gray'); plt.title('Бінарне зображення'); 
plt.show();
```

2.4 Встановити ядро розміром 2х2

```python
kernel = np.ones((2,2),np.uint8)
```

2.5 Застосувати ерозію до бінарного зображення та відобразити його

```python
erosion_st = cv2.erode(bin_img_st,kernel,iterations = 1)

plt.figure(figsize=(10,10))
plt.subplot(121);plt.imshow(erosion_st,cmap='gray'); plt.title('Ерозія'); 
plt.show();
```

2.6 Застосувати морфологічне відкриття до зображження після обробки ерозією та відобразити його

```python
opening = cv2.morphologyEx(erosion_st, cv2.MORPH_OPEN, kernel, iterations = 1)

plt.figure(figsize=(10,10));
plt.subplot(121);plt.imshow(opening, cmap='gray'); plt.title('Відкриття'); 
plt.show();
```

## 1. Методами порогової та морфологічної обробки відобразити тільки контури циферблату
 Завантажити зображення "digit.jpg" директорії "./examples/"
 
<table><tr>
<td> <img src="https://github.com/CryTechno/DIP/blob/main/examples/digit.jpg?raw=1" height="500" width="500"> </td>
</tr></table>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread("/content/digit.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap = 'gray', vmin = 0, vmax = 255);
plt.show()
bin_img = img.copy()
threshold =90
max_val = 255
min_val = 0
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val

kernel = np.ones((3,3),np.uint8)
gradient = cv2.morphologyEx(bin_img, cv2.MORPH_GRADIENT, kernel)


plt.imshow(gradient, cmap = 'gray', vmin = 0, vmax = 255);
plt.title('Контури циферблату'); 
```

## 2. Методами порогової та морфологічної обробки відобразити границі баскетбольного поля
 Завантажити зображення "basketball.jpg" директорії "./examples/"
 
<table><tr>
<td> <img src="https://github.com/CryTechno/DIP/blob/main/examples/basketball.jpg?raw=1" height="500" width="500"> </td>
</tr></table>

```python
img = cv2.imread("/content/basketball.jpg")
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
plt.imshow(img, cmap = 'gray', vmin = 0, vmax = 255);
img_trun = img.copy()
threshold = 250
max_val = 255
min_val = 0
ret, thresh_zero_inv = cv2.threshold(img, threshold, max_val, cv2.THRESH_TOZERO_INV)
plt.imshow(thresh_zero_inv, cmap = 'gray', vmin = 0, vmax = 255);
```

```python
threshold = 200

ret, thresh_zero = cv2.threshold(thresh_zero_inv, threshold, max_val, cv2.THRESH_TOZERO)
plt.imshow(thresh_zero, cmap = 'gray', vmin = 0, vmax = 255);
```

```python
kernel = np.ones((6,6),np.uint8)
erosion = cv2.erode(thresh_zero,kernel,iterations = 1)
plt.imshow(erosion, cmap = 'gray', vmin = 0, vmax = 255);
```
